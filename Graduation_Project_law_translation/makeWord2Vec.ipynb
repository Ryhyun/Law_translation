```python
from konlpy.tag import Mecab
import sys
import re
import codecs
from gensim.models import Word2Vec
from gensim.models.keyedvectors import KeyedVectors
from gensim.models import FastText as fText

from sklearn.cluster import KMeans
```

```python
test = []

pattern = "SSO|SSC|SC|SE|SW|SH|SY|SF|SN"
pos_tagger =Mecab()
nnpattern = "NNG|NNP"
def tokenize_basic(doc):
    a= []
    for t in pos_tagger.pos(doc):
        if( re.search(pattern, t[1]) != None):
            continue
        else:
            a.append('/'.join(t))
    return a

def tokenize_n(doc):
    pos_tagger =Mecab()

    a= []
    for t in pos_tagger.pos(doc):
        if( re.search(nnpattern, t[1]) != None):
            a.append('/'.join(t))
        else:
            continue;
    return a
```

```python
class SentenceReader:

    def __init__(self, filepath):
        self.filepath = filepath
        self.count = 0 ;

    def __iter__(self):
        for line in codecs.open(self.filepath, encoding='utf-8'):
            self.count += 1 
            if( self.count % 10000 == 0):
                print(self.count)
            yield tokenize_basic(line)
            
            
class SentenceReader_n:

    def __init__(self, filepath):
        self.filepath = filepath
        self.count = 0 ;

    def __iter__(self):
        for line in codecs.open(self.filepath, encoding='utf-8'):
            self.count += 1 
            if( self.count % 10000 == 0):
                print(self.count)
            yield tokenize_n(line)
```

```python
civil_law = SentenceReader("model/civil_law.txt")
real_estate_law =  SentenceReader("model/real_estate_law.txt")
traffic_law =SentenceReader("model/traffic.txt")


civil_law_n = SentenceReader_n("model/civil_law.txt")
real_estate_law_n =  SentenceReader_n("model/real_estate_law.txt")
traffic_law_n =SentenceReader_n("model/traffic.txt")
```

```python
db = DBSCAN(eps=5.5, min_samples=2).fit(civil_model_n[civil_model_n.wv.vocab])
core_samples_mask = np.zeros_like(db.labels_, dtype=int)
#core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
```

```python
print( labels)
print( n_clusters_)
cluster_20 = [[0]*0 for i in range(n_clusters_)]
print( len(labels))
```
